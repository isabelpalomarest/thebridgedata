### Step 1. Import the necessary libraries

```python

# remember to %matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

```python
chipo =  pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")
chipo
```

```python
chipo.head(10)
```

### Step 4. Create a histogram of the top 5 items bought

```python
# intento
by_item = chipo.groupby("item_name")["quantity"].count().sort_values(ascending=False).iloc[0:5]
by_item = pd.DataFrame(by_item)
hist=sns.displot(by_item)
hist.set(xlabel="item_name",ylabel="quantity")
by_item
```

```python
by_item.plot(kind ="bar")
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')
```

```python
by_item.plot()

# Set the title and labels
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')

# show the plot
plt.show()
```
