### Step 1. Import the necessary libraries

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
# remember to %matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
dicc = {"first_name": ["Jason", "Molly", "Tina", "Jake", "Amy"], "last_name": ["Miller", "Jacobson", "Ali", "Milner", "Cooze"], "age": [42, 52, 36, 24, 73], "female": [0, 1, 1, 0, 1], "preTestScore": [4, 24, 31, 2, 3],"postTestScore": [25, 94, 57, 62, 70]}
df = pd.DataFrame(dicc)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.scatterplot(x = df["preTestScore"], y = df["postTestScore"], s = df["age"])
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.scatterplot(x = df["preTestScore"], y = df["postTestScore"],hue= df["female"],s = df["postTestScore"]*4.5)
```
